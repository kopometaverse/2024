인공지능 모델 2024-03-20

```python
import tensorflow as tf
import pandas as pd
```

```python
#파일들로 부터 데이터 읽어오기

파일경로 = 'https://raw.githubusercontent.com/hyejeongcho/2024hitech/main/lemonade.csv'
레모네이드 = pd.read_csv(파일경로)
```

```python
# 변수안에 어떤 데이터가 있는지 확인

print(레모네이드.shape)
```

```python
# 칼럼 이름 출력

print(레모네이드.columns)
```

```python
# 독립 (영향을 주는 요소들) / 종송변수 (결과)
독립 = 레모네이드[['온도']]
종속 = 레모네이드[['판매량']]

print(독립.shape, 종속.shape)
```

```python
#실제데이터 출력

레모네이드.head()
```

```python
# 모델을 만듭니다.

X = tf.keras.layers.Input(shape=[1])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

```python
model.fit(독립,종속, epochs=2000)
```

```python
# 모델을 이용합니다.
model.predict([15])
```
